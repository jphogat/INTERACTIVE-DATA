```python
def convert_to_minutes(time_range):
    if 'More Than' in time_range:
        return 240  # Assuming "More Than 4 hour" is equivalent to 240 minutes
    else:
        hours, _, _ = time_range.partition('-')
        hours = int(hours.strip())
        return hours * 60
```

```python
# Apply the function to the column and create a new column
df['daily_studing_time_minutes'] = df['daily_studing_time'].apply(convert_to_minutes)
```

```python
def convert_social_medai_video_to_minutes(time_range):
    if 'More than' in time_range:
        return 180  # Assuming "more than 3 hour" is equivalent to 180 minutes
    else:
        parts = time_range.split('-')
        if len(parts) == 2:
            start, end = parts
            start_time, end_time = map(float, [start.strip().split()[0], end.strip().split()[0]])
            if 'hour' in time_range:
                return (end_time - start_time) * 60
            else:
                return end_time - start_time
        else:
            return float(time_range.split()[0])
```

```python
# Apply the function to the column and create a new column with minutes
df['social_media_Minutes'] = df['social_medai_&_video'].apply(convert_social_medai_video_to_minutes)
```

```python
def convert_travelling_time_to_minutes(time_range):
    if 'more than' in time_range:
        return 180  # Assuming "more than 3 hour" is equivalent to 180 minutes
    else:
        parts = time_range.split('-')
        if len(parts) == 2:
            start, end = parts
            start_time, end_time = map(float, [start.strip().split()[0], end.strip().split()[0]])
            if 'hour' in time_range:
                return (end_time - start_time) * 60
            else:
                return end_time - start_time
        else:
            return float(time_range.split()[0])
```

```python
# Apply the function to the column and create a new column with minutes
df['Travelling_Time_Minutes'] = df['Travelling_Time'].apply(convert_travelling_time_to_minutes)
```

```python
import statsmodels.api as sm
import pandas as pd

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Datsset.csv')
df.head()
# Assuming 'JoinedProgram' is binary (Yes/No)
df['JoinedProgram'] = df['JoinedProgram'].map({'Yes': 1, 'No': 0})

# Grouping the data by 'JoinedProgram' and calculating the mean of '10th_Mark' and '12th_Mark'
marks_by_program = df.groupby('JoinedProgram')[['10th Mark', '12th Mark']].mean().reset_index()

# Plotting the mean scores
plt.figure(figsize=(10, 6))
sns.barplot(x='JoinedProgram', y='10th Mark', data=marks_by_program, palette='viridis')
plt.title('Mean 10th Marks by JoinedProgram')
plt.xlabel('JoinedProgram')
plt.ylabel('Mean 10th Mark')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='JoinedProgram', y='12th Mark', data=marks_by_program, palette='viridis')
plt.title('Mean 12th Marks by JoinedProgram')
plt.xlabel('JoinedProgram')
plt.ylabel('Mean 12th Mark')
plt.show()
```

```python
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px

# Load the data
df = pd.read_csv('/content/Datsset.csv')
df.head()

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Student Data"),
    html.Div([
        dcc.Graph(id="table"),
        dcc.Graph(id="bar-chart")
    ])
])

# Define callback to update the table
@app.callback(
    Output("table", "figure"),
    [Input("bar-chart", "hoverData")]
)
def update_table(hoverData):
    if hoverData is None:
        selected_row = df.head(1)
    else:
        index = hoverData["points"][0]["pointIndex"]
        selected_row = df.iloc[[index]]

    fig = px.table(selected_row)
    fig.update_layout(margin=dict(t=20, l=20, r=20, b=20))
    return fig

# Define callback to update the bar chart
@app.callback(
    Output("bar-chart", "figure"),
    [Input("table", "hoverData")]
)
def update_bar_chart(hoverData):
    fig = px.bar(df, x="hobbies", y="salary expectation", color="Gender", barmode="group")
    fig.update_layout(margin=dict(t=20, l=20, r=20, b=20))
    return fig

# Run the app
if __name__ == "__main__":
    app.run_server(debug=True)
```

```python
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

# Load the data
df = pd.read_csv('/content/Datsset.csv')
df.head()
# Get unique hobbies
hobbies = df['hobbies'].unique()

# Initialize the Dash app
app = JupyterDash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Effect of Hobbies on Various Attributes"),
    html.Div([
        dcc.Dropdown(
            id='attribute-dropdown',
            options=[{'label': col, 'value': col} for col in df.columns],
            value='salary expectation',
            style={'width': '50%'}
        ),
        dcc.Dropdown(
            id='hobby-dropdown',
            options=[{'label': hobby, 'value': hobby} for hobby in hobbies],
            value='Video Games',
            style={'width': '50%'}
        ),
    ]),
    dcc.Graph(id="bar-chart")
])

# Define callback to update the bar chart
@app.callback(
    Output("bar-chart", "figure"),
    [Input("attribute-dropdown", "value"),
     Input("hobby-dropdown", "value")]
)
def update_bar_chart(attribute, hobby):
    fig = px.bar(df[df['hobbies'] == hobby], x=attribute, color="Gender", barmode="group",
                 title=f"Effect of {hobby} on {attribute}")
    fig.update_layout(margin=dict(t=50, l=20, r=20, b=20))
    return fig

# Run the app
app.run_server(mode='inline')
```

```python
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px


# Create the Dash app
app = dash.Dash(__name__)

# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("Student Dashboard"),
    html.Label("Select a feature to visualize:"),
    dcc.Dropdown(
        id='dropdown-feature',
        options=[{'label': col, 'value': col} for col in df.columns],
        value='Gender'
    ),
    html.Br(),
    dcc.Graph(id='bar-chart')
])

# Define callback to update the bar chart
@app.callback(
    Output('bar-chart', 'figure'),
    [Input('dropdown-feature', 'value')]
)
def update_bar_chart(selected_feature):
    if selected_feature:
        fig = px.bar(df, x=selected_feature, color="Gender", title=f"Distribution of {selected_feature}")
        return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

### Open a new Jupyter Notebook and import the pandas, matplotlib, and seaborn libraries in the same cell. Set the theme to "dark grid".

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
sns.set(style="whitegrid")
```

### Import the "Diabetes.csv Download Diabetes.csv" file and convert it to a DataFrame object using the pandas.read_csv() method.

```python
sea=pd.read_csv("Diabetes.csv")
sea.head()
```

```python
sns.scatterplot(x=sea["BMI"], y=sea["Glucose"])
plt.show()
```

```python
sns.relplot(data=sea, x="BMI",y="Glucose",hue="Pregnancies", size="Pregnancies", sizes=(10,50))
plt.xlim(10,47)
plt.ylim(37, 140)
plt.show()
```

### Create a scatter plot figure using the relplot function, with "Age" as the x-axis variable and "Insulin" as the y-axis variable.

```python
sns.relplot(data=sea, x="Age",y="Insulin", sizes=(10,50))
plt.show()
```

### In a new cell, create the same scatter plot figure, but this time add "Pregnancies" as the variable for "hue" and "size"

```python
sns.relplot(data=sea, x="Age",y="Insulin",hue="Pregnancies", size="Pregnancies", sizes=(10,50))
plt.xlim(10,40)
plt.ylim(20, 120)
plt.show()
```

### Finally, use the plt.xlim() and plt.ylim() methods to set limits for the x and y axes, and zoom in on your data

```python
sns.relplot(data=sea, x="Age",y="Insulin",hue="Pregnancies", size="Pregnancies", sizes=(10,50))
plt.xlim(10,40)
plt.ylim(20, 120)
plt.show()
```

```python
Sal=pd.read_csv("Salary Data.csv")
Sal.head()
```

### Create a scatter plot using Seaborn default settings with Age on the x-axis and Salary on the y-axis. Visualization:

```python
plt.Figure()
sns.scatterplot(x="Age",y="Salary", data=Sal)
plt.title("Default Scatter Plot : Age vs Salary")
plt.show()
```

### Explore the dataset and seaborn package by creating an advanced scatter plot; Include Gender and Education Level as attributes in the legend;

```python
plt.Figure()
sns.scatterplot(x="Age",y="Salary",hue="Education Level", style="Gender", data=Sal)
plt.title("Scatter Plot : Age vs Salary by education level and gender")
plt.show()
```

### Customize the color and marker palettes to increase visual appeal, Styling and Customization:


```python
education_colors={"Bachelor's":"Green", "Master's":"Blue", "PhD": "red"}
plt.Figure()
sns.scatterplot(x="Age",y="Salary",hue="Education Level", style="Gender", data=Sal, palette=education_colors)
plt.title("Scatter Plot : Age vs Salary by education level and gender")
plt.show()
```

```python
education_colors={"Bachelor's":"Green", "Master's":"Blue", "PhD": "red"}
gender_markers={'Female':"o", "Male":"s"}
plt.Figure()
sns.scatterplot(x="Age",y="Salary",hue="Education Level", style="Gender", data=Sal, palette=education_colors, markers=gender_markers)
plt.title("Scatter Plot : Age vs Salary by education level and gender")
plt.show()
```

### Set the plot style to remove grid lines for a cleaner look.


```python
sns.set_style("ticks")
sns.scatterplot(x="Age",
                y="Salary",
                hue="Education Level", 
                style="Gender", 
                data=Sal, 
                palette=education_colors,
                markers=gender_markers)
plt.legend(ncol=2, frameon=False, loc="lower right")
plt.title("Scatter Plot : Age vs Salary by education level and gender")
plt.show()
```

### Customize the legend to provide clear information without distracting from the plot. 

```python
gender_colors={"Female":"Green", "Male":"Red"}
plt.figure(figsize=(12,8))
sns.set_style("ticks")

Bachol=sns.scatterplot(x="Age",y="Salary",
                       hue="Education Level", 
                       style="Gender",
                       data=Sal, 
                       palette=education_colors, 
                       markers=gender_markers)
Bachol.set_xlim(25,50)
Bachol.set_ylim(25000,50000)
plt.legend(ncol=2, frameon=False, loc="lower left")
plt.title("Scatter Plot : Age vs Salary by education level and gender")
plt.show()
```

## 6. In a new code cell, create and show a figure that includes one axis with a bar plot of the `Evaluation` variable

```python
Evaluation_freq=cces.groupby("sex")["Evaluation"].count()
Evaluation_freq.index
fig6,ax6=plt.subplots()
plt.xlabel("Evaluation")
plt.ylabel("sex")
ax6.bar(x=Evaluation_freq.index, height=Evaluation_freq)
plt.show()
```

```python
plt.bar(Evaluation_freq.index, height=Evaluation_freq, color=['blue', 'red'])

# Add labels and title
plt.xlabel('sex')
plt.ylabel('Evaluation')
plt.title('Evaluation by sex')

# Show the plot
plt.show()
```

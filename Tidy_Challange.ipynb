```python
reaction=df[['timestamp_ms','reactions']]
# Explode the 'reactions' column
reaction = reaction.explode('reactions', ignore_index=True)

# Extract 'reaction' and 'actor' into separate columns
reaction[['reaction', 'actor']] = pd.json_normalize(reaction['reactions'])

# Drop the original 'reactions' column
reaction.drop(columns=['reactions'], inplace=True)

print(reaction)
```

```python
data_new=pd.merge(df, reaction,
how='left', on='timestamp_ms')
data_new.head()
```

Row-Level (Observation): Each row in the DataFrame represents a single message or post. Specifically, it contains information about a message or post sent by a user.

#### 5. In a Markdown cell, describe how the final DataFrame meets tidy data principles

##### The key principles of tidy data are as follows:

1. Each variable forms a column: In tidy data, each variable (i.e., an attribute, measurement, or characteristic) should be stored in its own column. This ensures that data attributes are clearly defined and easily accessible.

2. Each observation forms a row: Each observation (i.e., a single data point, instance, or record) should be stored in its own row. This allows for easy comparison and analysis of individual data points.

3. Each type of observational unit forms a table: Data should be organized into separate tables, with each table representing a distinct type of observational unit or entity. This helps maintain clear boundaries between different aspects of the data and avoids mixing unrelated information in the same table.

4. Columns contain atomic values: Columns should contain atomic, single-valued data elements. Avoid storing multiple values or complex data structures (such as lists or dictionaries) within a single column, as this can complicate data manipulation and analysis.

5. Column headers are informative and human-readable: Column headers (variable names) should be descriptive, concise, and human-readable. Use meaningful labels that clearly convey the content and context of the data.

The above datset meet all these five criteria of tidy principle.
